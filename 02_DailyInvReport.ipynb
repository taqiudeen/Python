```python
#import random
#import ctypes  # An included library with Python install. 
import pandas as pd
import mysql.connector
#from datetime import datetime, timedelta
from sqlalchemy import create_engine
engine = create_engine("mysql+pymysql://root:@localhost/taqwa_corp?charset=utf8mb4")


connect = mysql.connector.connect(
    host = "127.0.0.1",
    user = "root",
    password = "",
    database = "taqwa_corp"

)

mycursor = connect.cursor()
```

```python
#how many of the 62 parts are out of stock
queryOne = """
SELECT count(parts.PartNo) as OutOfStock
FROM `parts` 
WHERE parts.Inv <= 0 ORDER BY parts.Inv
"""
mycursor.execute(queryOne)
OutOfStock = mycursor.fetchone()
print("There are " + str(OutOfStock[0]) + " of the 62 parts out of stock. ")

#negative inventory with supplier orders
queryTwo = """
SELECT parts.PartNo, parts.Inv, worder.fullDate, worder.sO
FROM `parts` 
JOIN worder ON worder.partNo = parts.PartNo
WHERE parts.Inv <= 7 AND parts.pendO = 1 AND worder.recdInv = 0 ORDER BY parts.Inv
"""
dfOne = pd.read_sql(queryTwo, engine)
dfOne
```

```python
print("Sorted by Date")

#negative inventory with supplier orders, sorted by Date
querySix = """
SELECT parts.PartNo, parts.Inv, worder.fullDate, worder.sO
FROM `parts` 
JOIN worder ON worder.partNo = parts.PartNo
WHERE parts.Inv <= 7 AND parts.pendO = 1 AND worder.recdInv = 0 ORDER BY worder.fullDate
"""
dfFour = pd.read_sql(querySix, engine)
dfFour
```

```python
#how many of the 62 parts are out of stock and without re-orders
queryFour = """
SELECT count(parts.PartNo) as OutOfStock
FROM `parts` 
WHERE parts.pendO IS False AND parts.Inv <= 0 
"""

# the correct result is when pendO = 1 as opposed to 0 in the database, 
# the datatype is boolean, so you have to pass in zero for true and 1 for false, it seems 
# https://stackoverflow.com/questions/19431898/select-boolean-but-reverse-the-value

mycursor.execute(queryFour)
OutOfStock = mycursor.fetchone()
print("There are " + str(OutOfStock[0]) + " of the 62 parts out of stock witout re-orders. ")

#negative inventory without supplier orders
queryThree = """
SELECT partNo, Inv FROM `parts` WHERE Inv <= 0 AND pendO = 0 ORDER BY Inv
"""
dfTwo = pd.read_sql(queryThree, engine)
dfTwo
```

```python
#get last 7 before today
queryFive = """
SELECT ttlParts7Days
FROM admin
"""

mycursor.execute(queryFive)
LastSeven = mycursor.fetchone()

#LastSeven[0]
```

```python


#how many parts sold last week
querySix = """
SELECT count(invoices.PartNo) as TotalSold
FROM invoices
JOIN orders ON orders.OID = invoices.OID
WHERE orders.oDate >= CURRENT_DATE() - 7 AND orders.oDate < CURRENT_DATE() 
"""

mycursor.execute(querySix)
OutOfStock = mycursor.fetchone()
print("There were " + str(OutOfStock[0]) + " parts sold in the last 7 days. ")
print ("There was a difference of " + str(LastSeven[0] - OutOfStock[0]) + " parts in the last 7 days.")

sql = "UPDATE admin SET ttlParts7Days = %s"
#val = [OutOfStock]
mycursor.execute(sql, OutOfStock)  
connect.commit()

#how many of each part sold last week
querySeven = """
SELECT invoices.PartNo, count(invoices.PartNo) as TotalSold
FROM invoices
JOIN orders ON orders.OID = invoices.OID
WHERE orders.oDate >= CURRENT_DATE() - 7 AND orders.oDate < CURRENT_DATE() 
GROUP BY invoices.PartNo
ORDER BY TotalSold DESC
"""

dfThree = pd.read_sql(querySeven, engine)
dfThree
```

```python
mycursor.close()
connect.close()
```
