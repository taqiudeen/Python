```python
import pandas as pd
#import mysql.connector
#from datetime import datetime, timedelta
from sqlalchemy import create_engine
engine = create_engine("mysql+pymysql://root:@localhost/d_survey?charset=utf8mb4")
```

```python
#negative inventory with supplier orders
queryOne = """
SELECT *
FROM survey_1
"""
dfSurveyOne = pd.read_sql(queryOne, engine)
dfSurveyOne
```

```python
#
queryOne = """
SELECT *
FROM survey_2
"""
dfSurveyTwo = pd.read_sql(queryOne, engine)
dfSurveyTwo
```

```python
dfAvgResults = pd.DataFrame(columns=['Question', 'Survey 1', 'Survey 2', 'Dif'])
dfAvgResults.set_index('Question', inplace=True)
dfAvgResults
```

```python
i = 1

while i <= 9:
    qNo = "Q" + str(i)
    new_row = {'Question': i, 'Survey 1': dfSurveyOne[qNo].mean(), 'Survey 2': dfSurveyTwo[qNo].mean(), 'Dif': dfSurveyTwo[qNo].mean() - dfSurveyOne[qNo].mean()}
    dfAvgResults.loc[i] = new_row
    i = i + 1
    
dfAvgResults
```
